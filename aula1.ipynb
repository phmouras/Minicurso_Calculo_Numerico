```python
# Derivada numérica: Exemplo 1, função seno

x = 0
delta_x = 1e-5

# Define derivada de y por x pela definição
y_x = ( math.sin(x+delta_x) - math.sin(x)) / delta_x

print(y_x)
```

```python
# Construíndo o gráfico em um domínio pré-definido

# x é um vetor de posições de 0 a 10, igualmente espaçado em 100 elementos
x = np.linspace(0,10,100)

# Define delta_x para ser aplicado em cada ponto de x
delta_x = 1e-5


# Calcula a derivada em todos os pontos
y_x = (np.sin(x+delta_x) - np.sin(x)) / delta_x



# Esquema de plotagem para apresentar a derivada em todos os pontos
plt.figure(figsize=(6, 6))     
    
plt.xlabel("x")
plt.ylabel("y(x)")
plt.plot(x, y_x, 'x', markersize=2, label='y_x')
plt.plot(x, np.cos(x), '-', markersize=2, label='cos(x)')
plt.grid()
plt.legend(loc="upper right")
plt.show()
plt.clf()
```

```python
# Derivada numérica: Exemplo 2, função logarítmica

x = 1
delta_x = 1e-5


y_x = ( math.log(x+delta_x) - math.log(x)) / delta_x

print(y_x)
```

```python
# Construíndo o gráfico em um domínio pré-definido

# x é um vetor de posições de 0 a 10, igualmente espaço em 100 elementos
x = np.linspace(1,10,100)
delta_x = 1e-5

y_x = (np.log(x+delta_x) - np.log(x)) / delta_x

# Esquema de plotagem para apresentar a derivada em todos os pontos
plt.figure(figsize=(6, 6))     
    
plt.xlabel("x")
plt.ylabel("y(x)")
plt.plot(x, y_x, 'x', markersize=2, label='y_x')
plt.plot(x, 1/x, '-', markersize=2, label='1/x')
plt.grid()
plt.legend(loc="upper right")
plt.show()
plt.clf()
```

```python
# Integração numérica: Exemplo 1, função quadrática

n_pontos = 1000
x = np.linspace(0,1,n_pontos)
delta_x = x[1]-x[0]
y = 0


for i in range(len(x)):
    
    y = y + delta_x * ( x[i]**2 + x[i] + 2 )


print('Integral numérica: ', y)
print('Resultado analítico: ', 1/3+1/2+2)
```

```python
# Integração numérica: Exemplo 2, função exponencial

n_pontos = 100
x = np.linspace(0,1,n_pontos)
delta_x = 1.0/n_pontos
y = 0

for i in range(len(x)):
    
    y = y + delta_x * x[i]**x[i]


print('Integral numérica: ', y)
print('Resultado Wolfram:  0.783431...')
```

```python
# Método de Euler: Resfriamento de Newton

n_pontos = 100
t = np.linspace(0,5,n_pontos)
delta_t = 5.0/n_pontos
k = 2
T_a = 1
T_0 = 10

T = [T_0]

for i in range(len(t)-1):
    
    y = T[i] - k * delta_t * (T[i] - T_a)
    T.append(y)



plt.figure(figsize=(6, 6))     
    
plt.xlabel("t")
plt.ylabel("T(t)")
plt.plot(t, T, 'x', markersize=2, label='Euler')
plt.plot(t, (T_0-T_a)*np.exp(-k*t)+T_a, '-', markersize=2, label='Analítico')

plt.grid()
plt.legend(loc="upper right")
plt.show()
plt.clf()
```

```python
# Método de Euler: Oscilador Harmônico

n_pontos = 1000
t = np.linspace(0,5,n_pontos)
delta_t = 5.0/n_pontos
omega2 = 4
x_0 = 1
v_0 = 0

x = [x_0]
v = [v_0]

for i in range(len(t)-1):
    
    y = x[i] +  delta_t * v[i]
    x.append(y)
    
    y = v[i] - omega2 * delta_t * x[i]                # x[i] -----> x[i+1] para usar Euler-Cramer
    v.append(y)
    



plt.figure(figsize=(6, 6))     
    
plt.xlabel("t")
plt.ylabel("x(t)")
plt.plot(t, x, 'x', markersize=1, label='Euler')
plt.plot(t, np.cos(math.sqrt(omega2)*t), '-', markersize=2, label='Analítico')

plt.grid()
plt.legend(loc="upper right")
plt.show()
plt.clf()
```

```python
# Método de Euler: Pêndulo simples

n_pontos = 1000
t = np.linspace(0,5,n_pontos)
delta_t = 5.0/n_pontos
omega2 = 4
x_0 = 1
v_0 = 0

x = [x_0]
v = [v_0]

for i in range(len(t)-1):
    
    y = x[i] +  delta_t * v[i]
    x.append(y)
    
    y = v[i] - omega2 * delta_t * np.sin(x[i])                # x[i] -----> x[i+1] para usar Euler-Cramer
    v.append(y)
    



plt.figure(figsize=(6, 6))     
    
plt.xlabel("t")
plt.ylabel("x(t)")
plt.plot(t, x, 'x', markersize=1, label='Euler')
plt.plot(t, np.cos(math.sqrt(omega2)*t), '-', markersize=2, label='Analítico')

plt.grid()
plt.legend(loc="upper right")
plt.show()
plt.clf()
```
